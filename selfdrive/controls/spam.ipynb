```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 6]
plt.rcParams['figure.dpi'] = 100
import numpy as np

desired = []
latplan = []
with open("dump.txt", "r") as f:
  for line in f:
    l = line.strip().split()
    if l[0]=="a":
      desired.append(float(l[1]))
    elif l[0]=="b":
      latplan.append(float(l[1]))
l = 6000
diff = 2000
des2 = desired[l:l+diff]
offset =20
lat2 = latplan[l+offset:l+diff+offset]


plt.plot(np.linspace(0, len(des2)/100, len(des2)), des2, label="desired")
plt.plot(np.linspace(0, len(lat2)/100, len(lat2)), lat2, label="latplan")
```
